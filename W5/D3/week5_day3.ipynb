```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
import ipywidgets as widgets

# Initial data
t = np.linspace(0, 1, 1000)
initial_freq = 5

def plot_sine_wave(freq):
    y = np.sin(2 * np.pi * freq * t)
    fig, ax = plt.subplots()
    ax.plot(t, y)
    ax.set_ylim(-1.1, 1.1)
    plt.show()

# Slider
freq_slider = FloatSlider(value=initial_freq, min=0.1, max=30.0, step=0.1, description='Freq')

# Interactive plot
interact(plot_sine_wave, freq=freq_slider)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate three datasets with normal distributions
data1 = np.random.normal(100, 10, 200)  # Mean=100, Std=10, Size=200
data2 = np.random.normal(90, 20, 200)   # Mean=90, Std=20, Size=200
data3 = np.random.normal(80, 30, 200)   # Mean=80, Std=30, Size=200

# Combine the datasets into a list for the box plot
data = [data1, data2, data3]

# Create a figure with a specific size
fig = plt.figure(figsize=(10, 7))  # Set figure dimensions to 10x7 inches
ax = fig.add_subplot(111)  # Add a single subplot to the figure

# Create the box plot
bp = ax.boxplot(data, patch_artist=True, notch=True, vert=0)
# Parameters:
# - patch_artist=True: Fills the boxes with color.
# - notch=True: Adds a notch to indicate the confidence interval of the median.
# - vert=0: Makes the box plot horizontal (default is vertical).

# Customize the appearance of the whiskers
for whisker in bp['whiskers']:
    whisker.set(color='yellow', linewidth=1.5, linestyle=":")
    # - color='yellow': Sets the whisker color.
    # - linewidth=1.5: Sets the thickness of the whisker lines.
    # - linestyle=":": Makes the whisker lines dotted.

# Customize the appearance of the caps (end lines of whiskers)
for cap in bp['caps']:
    cap.set(color='red', linewidth=2)
    # - color='red': Sets the cap color.
    # - linewidth=2: Sets the cap line thickness.

# Customize the appearance of the median lines
for median in bp['medians']:
    median.set(color='green', linewidth=3)
    # - color='green': Sets the median line color.
    # - linewidth=3: Sets the median line thickness.

# Customize the style of the fliers (outlier points)
for flier in bp['fliers']:
    flier.set(color='#e7298a', alpha=0.7)
    # - color='#e7298a': Sets the color of the outliers.
    # - alpha=0.7: Makes the outliers slightly transparent.

# Add a title to the box plot
plt.title('Customized Box Plot')  # Sets the title for the plot

# Display the box plot
plt.show()  # Renders the plot
```

```python
import pandas as pd
import numpy as np
from plotnine import *  # Import all plotnine functions

# Set random seed for reproducibility
np.random.seed(42)

# Dataset with homoscedasticity (constant variance)
x_homo = np.linspace(1, 10, 200)
y_homo = 2 * x_homo + np.random.normal(0, 2, 200)
data_homo = pd.DataFrame({'x': x_homo, 'y': y_homo})

# Dataset with heteroscedasticity (increasing variance)
x_hetero = np.linspace(1, 10, 200)
y_hetero = 2 * x_hetero + np.random.normal(0, 0.2 * x_hetero, 200)
data_hetero = pd.DataFrame({'x': x_hetero, 'y': y_hetero})

# Dataset with more data points in certain areas and a non-linear trend
x_dense = np.concatenate([np.random.uniform(1, 3, 100),
                         np.random.uniform(3, 7, 50),
                         np.random.uniform(7, 10, 150)])
y_dense = 2 * x_dense + 0.5 * x_dense**2 + np.random.normal(0, 5, 300)
data_dense = pd.DataFrame({'x': x_dense, 'y': y_dense})

# Plotting function
def create_plot(data, title, x_lab, y_lab):
    return (ggplot(data, aes(x='x', y='y', color='x')) +
            geom_point(alpha=0.6) +
            geom_smooth(method='lm', color='red', se=True) +
            theme_minimal() +
            theme(figure_size=(10, 6)) +  # Add figure size control
            ggtitle(title) +
            labs(x=x_lab, y=y_lab) +
            scale_color_gradient(low="blue", high="green"))

# Create and display plots
plot_homo = create_plot(data_homo,
                       'Homoscedasticity (Constant Variance)',
                       'X values',
                       'Y values')
plot_homo.show()

plot_hetero = create_plot(data_hetero,
                         'Heteroscedasticity (Increasing Variance)',
                         'X values',
                         'Y values')
plot_hetero.show()

plot_dense = create_plot(data_dense,
                        'Non-linear Trend with Varying Data Density',
                        'X values',
                        'Y values')
plot_dense.show()
```

```python
# First, install plotnine if you haven't already
!pip install plotnine

import pandas as pd
import numpy as np
from plotnine import *  # Import all plotnine functions

# Set random seed for reproducibility
np.random.seed(42)

# Generate example data
date_range = pd.date_range(start='2021-01-01', periods=100, freq='D')
temperature = np.random.normal(loc=20, scale=5, size=100)
humidity = np.random.normal(loc=60, scale=10, size=100)
category = np.random.choice(['A', 'B', 'C'], size=100)
data = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity,
    'Category': category
})

# Scatter Plot: Temperature vs. Humidity
scatter_plot = (ggplot(data, aes(x='Temperature', y='Humidity', color='Category')) +
                geom_point(size=3, alpha=0.6) +
                geom_smooth(method='lm', se=False) +
                theme_minimal() +
                theme(figure_size=(10, 6)) +
                labs(title="Scatter Plot: Temperature vs. Humidity",
                     subtitle="Colored by Category",
                     x="Temperature (°C)",
                     y="Humidity (%)",
                     color="Category") +
                scale_color_manual(values=["#1f77b4", "#2ca02c", "#d62728"]))
scatter_plot.show()

# Line Plot: Temperature over Time
line_plot = (ggplot(data, aes(x='Date', y='Temperature', color='Category')) +
             geom_line(size=1) +
             theme_minimal() +
             theme(figure_size=(10, 6)) +
             labs(title="Line Plot: Temperature over Time",
                  subtitle="Colored by Category",
                  x="Date",
                  y="Temperature (°C)",
                  color="Category") +
             scale_color_manual(values=["#1f77b4", "#2ca02c", "#d62728"]))
line_plot.show()

# Faceted Histogram: Temperature Distribution by Category
histogram_plot = (ggplot(data, aes(x='Temperature', fill='Category')) +
                 geom_histogram(binwidth=1, alpha=0.7, position='identity') +
                 facet_wrap('~Category', ncol=3) +
                 theme_minimal() +
                 theme(figure_size=(12, 4),
                       subplots_adjust={'wspace': 0.2}) +
                 labs(title="Faceted Histogram: Temperature Distribution by Category",
                      subtitle="Each panel represents a different category",
                      x="Temperature (°C)",
                      y="Count",
                      fill="Category") +
                 scale_fill_manual(values=["#1f77b4", "#2ca02c", "#d62728"]))
histogram_plot.show()
```

```python
import pandas as pd
from plotnine import ggplot, aes, geom_line, geom_bar, geom_point, geom_histogram, geom_smooth, labs, theme_minimal, scale_fill_manual

# Load the dataset from the main folder in Google Colab
file_path = '/content/Cleaned_Students_Performance.csv'
student_data = pd.read_csv(file_path)

# Ensure 'Date' columns or any non-numeric values are not blocking
print(student_data.head())

# 1. Line plot: Average scores across parental education levels
line_plot = (
    ggplot(student_data, aes(x='parental_level_of_education', y='average_score', group=1)) +
    geom_line(stat='summary', fun_y='mean', color='blue', size=1.5) +
    labs(
        title='Average Scores by Parental Education Level',
        x='Parental Level of Education',
        y='Average Score'
    ) +
    theme_minimal()
)

# 2. Bar plot: Comparing average scores by gender
bar_plot = (
    ggplot(student_data, aes(x='factor(gender)', y='average_score', fill='factor(gender)')) +
    geom_bar(stat='summary', fun_y='mean', position='dodge') +
    scale_fill_manual(values=['blue', 'orange'], labels=['Female', 'Male']) +
    labs(
        title='Average Scores by Gender',
        x='Gender',
        y='Average Score',
        fill='Gender'
    ) +
    theme_minimal()
)

# 3. Scatter plot: Relationship between math and reading scores
scatter_plot = (
    ggplot(student_data, aes(x='math_score', y='reading_score')) +
    geom_point(alpha=0.6, color='green') +
    geom_smooth(method='lm', color='red') +
    labs(
        title='Relationship Between Math and Reading Scores',
        x='Math Score',
        y='Reading Score'
    ) +
    theme_minimal()
)

# 4. Histogram: Distribution of average scores with KDE
histogram_plot = (
    ggplot(student_data, aes(x='average_score')) +
    geom_histogram(binwidth=5, fill='blue', alpha=0.7) +
    geom_smooth(aes(y='..density..'), method='loess', se=False, color='red', linetype='dashed') +
    labs(
        title='Distribution of Average Scores',
        x='Average Score',
        y='Count'
    ) +
    theme_minimal()
)

# Display the plots
line_plot.show()
bar_plot.show()
scatter_plot.show()
histogram_plot.show()
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# URL of the Earthquake Database
url = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(url)

# Select relevant columns for plotting
df = df[['place', 'latitude', 'longitude', 'mag']]

# Display the first few rows to verify the selection
print(df.head())

# Create a scatter geo plot for earthquakes
fig = px.scatter_geo(df,
                     lat='latitude',
                     lon='longitude',
                     hover_name='place',
                     color='mag',
                     title='All Earthquakes (Past Month)',
                     color_continuous_scale=px.colors.cyclical.IceFire)

# Update the layout of the map
fig.update_layout(
    geo=dict(
        showland=True,
        landcolor='rgb(212, 212, 212)',
    )
)

# Show the plot
fig.show()

# Create a density mapbox for earthquakes
heatmap_fig = go.Figure(go.Densitymapbox(lat=df.latitude, lon=df.longitude, z=[1]*len(df),
                                         radius=10, showscale=True))

# Set the layout for the mapbox
heatmap_fig.update_layout(mapbox_style="stamen-terrain",
                          mapbox_center_lat=0,
                          mapbox_center_lon=0,
                          mapbox_zoom=1,
                          title='Heatmap of All Earthquakes (Past Month)')

# Show the heatmap
heatmap_fig.show()
```

```python
import requests
import zipfile
import io
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Ensure the output is displayed in the notebook
output_notebook()

# Create sample fish dataset (since we can't directly download from Kaggle)
# This creates a synthetic dataset similar to the fish dataset structure
np.random.seed(42)

# Create synthetic data
n_samples = 1000
species = ['Bream', 'Roach', 'Whitefish', 'Parkki', 'Perch', 'Pike', 'Smelt']
data = {
    'Species': np.random.choice(species, n_samples),
    'Length1': np.random.normal(25, 10, n_samples),
    'Weight': []
}

# Create correlated weight values
for length in data['Length1']:
    # Weight is roughly proportional to length^3 with some noise
    weight = 0.01 * (length ** 3) + np.random.normal(0, 20)
    data['Weight'].append(max(weight, 0))  # Ensure non-negative weights

df = pd.DataFrame(data)

# Sample a fraction (10%) of the data
sampled_df = df.sample(frac=0.1, random_state=42)

# Hexbin plot using Pandas and Matplotlib
plt.figure(figsize=(10, 6))
plt.hexbin(sampled_df['Length1'], sampled_df['Weight'], gridsize=50, cmap='viridis')
plt.title('Hexbin Plot: Length1 vs Weight')
plt.xlabel('Length1')
plt.ylabel('Weight')
plt.colorbar(label='Count')
plt.show()

# Create a basic interactive plot with Bokeh
p = figure(title="Interactive Scatter Plot: Length1 vs Weight with Bokeh",
           x_axis_label='Length1',
           y_axis_label='Weight')
p.circle(sampled_df['Length1'], sampled_df['Weight'], size=2, alpha=0.5)

# Show the plot
show(p)

# Scatter plot using Seaborn with transparency to manage overplotting
plt.figure(figsize=(10, 6))
sns.scatterplot(data=sampled_df, x='Length1', y='Weight', hue='Species', alpha=0.3, palette='viridis')
plt.title('Scatter Plot with Transparency: Length1 vs Weight')
plt.xlabel('Length1')
plt.ylabel('Weight')
plt.show()

# Add jittering
sampled_df['Length1_jitter'] = sampled_df['Length1'] + np.random.normal(0, 0.1, size=len(sampled_df))
sampled_df['Weight_jitter'] = sampled_df['Weight'] + np.random.normal(0, 0.1, size=len(sampled_df))

# Scatter plot with jittering using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=sampled_df, x='Length1_jitter', y='Weight_jitter', hue='Species', alpha=0.3, palette='viridis')
plt.title('Scatter Plot with Jittering: Length1 vs Weight')
plt.xlabel('Length1')
plt.ylabel('Weight')
plt.show()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

# Generate random data for points on the globe
num_points = 100
latitudes = np.random.uniform(-90, 90, num_points)  # Random latitudes between -90 and 90
longitudes = np.random.uniform(-180, 180, num_points)  # Random longitudes between -180 and 180
magnitudes = np.random.uniform(1, 10, num_points)  # Random "sizes" for points

# Create a DataFrame to manage data
df = pd.DataFrame({
    'latitude': latitudes,
    'longitude': longitudes,
    'magnitude': magnitudes
})

# Create a scattergeo plot for 3D spinning map
fig = go.Figure()

# Add scattergeo layer
fig.add_trace(go.Scattergeo(
    lon=df['longitude'],  # Longitude for points
    lat=df['latitude'],  # Latitude for points
    text=df['magnitude'],  # Hover text to display magnitude
    mode='markers',  # Markers only
    marker=dict(
        size=df['magnitude'],  # Scale marker size by "magnitude"
        color=df['magnitude'],  # Color based on magnitude
        colorscale='Viridis',  # Color scale
        cmin=1,  # Minimum color scale value
        cmax=10,  # Maximum color scale value
        colorbar=dict(title="Magnitude"),  # Add colorbar with a title
    )
))

# Set layout for the globe
fig.update_geos(
    projection_type="orthographic",  # 3D globe projection
    showland=True,  # Display land
    landcolor="rgb(240, 240, 240)",  # Land color
    oceancolor="rgb(0, 100, 255)",  # Ocean color
    showocean=True,  # Display ocean
    showcountries=True,  # Show country boundaries
    countrycolor="rgb(200, 200, 200)"  # Country boundary color
)

# Add spinning effect and title
fig.update_layout(
    title="3D Interactive Spinning World Map",
    margin=dict(l=0, r=0, t=40, b=0),  # Reduce margins for better visualization
    geo=dict(bgcolor='rgba(0,0,0,0)'),  # Transparent background
)

# Make the map rotate automatically
fig.update_layout(updatemenus=[
    dict(
        type="buttons",
        showactive=False,
        buttons=[
            dict(label="Play",
                 method="animate",
                 args=[None, dict(frame=dict(duration=50, redraw=True), fromcurrent=True)])
        ]
    )
])

# Add animation frames for rotation
frames = []
for lon in np.linspace(0, 360, 180):
    frames.append(go.Frame(layout=dict(
        geo=dict(projection_rotation=dict(lon=lon))
    )))

fig.frames = frames

# Show the plot
fig.show()
```
