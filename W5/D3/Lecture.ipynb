# Basic Plotting with Plotnine
Creating a Simple Plot
- Hereâ€™s an example of creating a basic scatter plot using Plotnine:

```python
from plotnine import ggplot, aes, geom_point
import pandas as pd

# Sample data
data = pd.DataFrame({'x': [1, 2, 3, 4, 5], 'y': [3, 5, 2, 6, 4]})

# Creating a scatter plot
(ggplot(data, aes(x='x', y='y')) + geom_point())
```

```python
# Advanced Plotting Techniques

# Layering in Plotnine
# With Plotline, you can add multiple layers to a plot for enhanced visualization.
# Adding Layers
# To add a layer, such as a smooth line on top of a scatter plot:


from plotnine import geom_smooth

# Adding a smooth line layer
(ggplot(data, aes(x='x', y='y')) + 
 geom_point() + 
 geom_smooth(method='lm'))
```

```python
import pandas as pd
import numpy as np
from plotnine import ggplot, aes, geom_point, scale_color_manual, ggtitle
np.random.seed(42)

# Generate example data
x = np.linspace(1, 10, 200)
y = 2 * x + np.random.normal(0, 2, 200)

# Create a categorical variable
category = np.random.choice(['A', 'B', 'C'], size=200)

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y, 'category': category})

# Plot with customized colors
plot = (ggplot(data, aes(x='x', y='y', color='category')) +
        geom_point() +
        scale_color_manual(values=["blue", "green", "red"]) +
        ggtitle("Example Plot"))

# Display the plot
plot.show()
```

```python
# Using Themes
from plotnine import theme_bw

(ggplot(data, aes(x='x', y='y')) + 
 geom_point() + 
 theme_bw())
```
