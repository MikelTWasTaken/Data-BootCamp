```python
# ðŸŒŸ Exercise 7: Data Transformation for Age Feature
# Instructions
# Create age groups (bins) from the Age column to categorize passengers into different age categories.
# Apply one-hot encoding to the age groups to convert them into binary features.
# Hint: Use pd.cut() for binning the Age column and pd.get_dummies() for one-hot encoding.

import pandas as pd

age_bins = [0, 12, 18, 35, 60, 100]  # Define bin edges
age_labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']  # Define bin labels
titanic_data['Age_Group'] = pd.cut(titanic_data['Age'], bins=age_bins, labels=age_labels, right=False)

print("DataFrame with Age Groups:")
print(titanic_data)

titanic_data = pd.get_dummies(titanic_data, columns=['Age_Group'], prefix='AgeGroup', drop_first=False)

print("\nDataFrame with One-Hot Encoded Age Groups:")
print(titanic_data)



```
