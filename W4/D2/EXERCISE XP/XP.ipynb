```python
# ðŸŒŸExercise 5: Basic Data Analysis with Google Colab


# Objective: Perform basic data analysis using Google Colab.
# Task:
# Using the same notebook from the previous exercise, perform basic data analysis tasks:
# Calculate the mean, median, and mode of a quantitative column.
mean_PetalWidthCm = Iris['PetalWidthCm'].mean()
print(f"Mean PetalWidthCm: {mean_PetalWidthCm}")

median_SepalLengthCm = Iris['SepalLengthCm'].median()
print(f"Median SepalLengthCm: {median_SepalLengthCm}")

mode_SepalWidthCm = Iris['SepalWidthCm'].mode()
print(f"Mode SepalWidthCm: {mode_SepalWidthCm}")

# Create a simple plot (like a histogram or bar chart) to visualize the data using Matplotlib/Seaborn libraries :
import matplotlib.pyplot as plt
data = pd.read_csv('/Users/teitelbaumsair/Downloads/Iris Dataset/Iris.csv')
plt.bar(data['SepalWidthCm'], data['SepalLengthCm'], color='skyblue')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Sepal Length (cm)')
plt.title('Sepal Width vs. Sepal Length')
plt.show()

# Document your findings in the notebook.
#Generally speaking Sepal's with an average width are Longer. slim and wide sepals are generally shorter. 
# Tools: Google Colab, Python with Pandas and Matplotlib/Seaborn libraries.
```
