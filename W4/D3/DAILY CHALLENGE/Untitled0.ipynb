```python
# Load an image using PIL
image_path = 'flowers/flowers/19_010.png'
original_image = Image.open(image_path)

# Display the original image using matplotlib
plt.imshow(original_image)
```

```python
rotated_image = original_image.rotate(50)
plt.imshow(rotated_image)
```

```python
from PIL import Image,ImageOps
mirror_image = ImageOps.mirror(original_image)
plt.imshow(mirror_image)
```

```python
from PIL import Image,ImageOps
flipped_image = ImageOps.flip(original_image)
plt.imshow(flipped_image)
```

```python
from PIL import Image,ImageOps
new_size = (1000, 1200)
zoom_image = original_image.resize(new_size)
plt.imshow(zoom_image)
```
