```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from scipy.ndimage import rotate
import numpy as np
import matplotlib.pyplot as plt

# Function to rotate image by 30 degrees
def rotate_image_30_degrees(image):
    return rotate(image, 30, reshape=False, mode='nearest')

# Load the image
image = load_img('cat.jpg', target_size=(234, 234))  # Load and resize image
image_array = img_to_array(image)  # Convert the image to an array

# Rotate the image by 30 degrees
rotated_image = rotate_image_30_degrees(image_array)

# Plot the original and rotated images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image_array.astype('uint8'))
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(rotated_image.astype('uint8'))
plt.title("Rotated Image (30)")

plt.show()
```
