Step 5: Ask a question about the Data
- try to answer a question you have about the plot or statistic.

What are the locations with the fastest rollercoasters(minimum of 10)?


```python
ax = df.query('Location != "Other"') \
    .groupby('Location')['Speed_MPH'] \
    .agg(['mean','count']) \
    .query('count >= 10')\
    .sort_values('mean')['mean']\
    .plot(kind='barh', figsize=(12, 5), title='Average Speed by Location (MPH)')
ax.set_xlabel('Average Speed')
```
