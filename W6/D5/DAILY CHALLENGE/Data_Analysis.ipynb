# Hypothesis Testing:

### Three G vs. Four G:
- **H0 (Null)**: There is no significant difference in the distribution or relationship between the presence of 3G (`three_g`) and 4G (`four_g`) features in the dataset.
- **H1 (Alt)**: There is a significant difference in the distribution or relationship between the presence of 3G (`three_g`) and 4G (`four_g`) features in the dataset.

### Fc (Front Camera) vs. Price:
- **H0 (Null)**: The front camera feature (`fc`) does not have a significant relationship with the price of the phone.
- **H1 (Alt)**: The front camera feature (`fc`) has a significant relationship with the price of the phone.




### Three G vs. Four G:

```python
import pandas as pd
from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(df['three_g'], df['four_g'], equal_var=False)  # assuming unequal variance

# Print results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Hypothesis testing
alpha = 0.05  # significance level

if p_value < alpha:
    print("\nReject the null hypothesis (H0): There is a significant difference between 3G and 4G features.")
else:
    print("\nFail to reject the null hypothesis (H0): There is no significant difference between 3G and 4G features.")
```

### Fc (Front Camera MGP) vs. Price:

To conduct a statistically accurate understanding of FC on price, eventually we have to look at the dataset and correlation of phones with a camera (indicated by a value greataer than 0 in the column and 0 meaning there are no megapixels ergo, no front camera. Oppositley, anything with a number indicates the presence of a front camera). Then we can compare prices of phones without a front camera to those with and see if there is any significance to the camera on price at all. 

```python
has_front_camera = df['fc'] > 0
has_no_fc = df['fc'] == 0

# Phones with a front camera
phones_with_fc = df[has_front_camera]

# Phones without a front camera
phones_without_fc = df[has_no_fc]

# Calculate the average price for each group
avg_price_with_fc = phones_with_fc['price_range'].mean()
avg_price_without_fc = phones_without_fc['price_range'].mean()

print("Avg price of phones with a front camera: ", avg_price_with_fc)
print("Avg price of phones without a front camera: ", avg_price_without_fc)
```

```python
import scipy.stats as stats
import pandas as pd

# Calculate the Pearson correlation coefficient between fc and price
correlation, p_value = stats.pearsonr(df['fc'], df['price_range'])

# Print the results
print(f"Pearson Correlation: {correlation}")
print(f"P-value: {p_value}")

# Hypothesis Testing
alpha = 0.05  # significance level

if p_value < alpha:
    print("Reject the null hypothesis (H1: There is a significant relationship between fc and price).")
else:
    print("Fail to reject the null hypothesis (H0: There is no significant relationship between fc and price).")
```
