```python
import scipy.stats as stats
import numpy as np
# Given a sample dataset, calculate the mean, median, variance, and standard deviation using SciPy.
# Sample dataset in code:
data = [12, 15, 13, 12, 18, 20, 22, 21]


mean_scipy = stats.tmean(data)
median_scipy = np.median(data)  # SciPy does not have a separate median function
variance_scipy = stats.tvar(data)
std_dev_scipy = stats.tstd(data)

print(f"Mean: {mean_scipy}")
print(f"Median: {median_scipy}")
print(f"Variance: {variance_scipy}")
print(f"Standard Deviation: {std_dev_scipy}")
```

```python
# Generate a normal distribution using SciPy with a mean of 50 and a standard deviation of 10. Plot this distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 50
std_dev = 10

x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
y = norm.pdf(x, mean, std_dev)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f"Mean = {mean}, Std Dev = {std_dev}")
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()
plt.show()
```

```python
# Perform a T-test on two sets of randomly generated data. Hint: Use numpy.random.normal() to generate the datasets.

import numpy as np
from scipy import stats

df = np.random.uniform(size=1000)
df2 = np.random.uniform(size=1000)
t_stat, p_value = stats.ttest_rel(df, df2)
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
t_stat, p_value = stats.ttest_ind(df, df2)
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
# Task: Import a dataset using Pandas and calculate descriptive statistics for a specific column using SciPy.
# Sample Dataset: Use any publicly available dataset (e.g., Iris dataset).
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
iris["area_proportion"] = 0.0

for species in iris["species"].unique():
    species_data = iris[iris["species"] == species]
    individual_area = np.multiply(species_data["sepal_length"], species_data["sepal_width"])
    overall_area = np.sum(individual_area)
    iris.loc[iris["species"] == species, "area_proportion"] = individual_area / overall_area
plt.figure(figsize=(12, 6))
for species in iris["species"].unique():
    species_data = iris[iris["species"] == species]
    plt.scatter(
        species_data["sepal_length"],
        species_data["sepal_width"],
        s=species_data["area_proportion"] * 10000,  # Scale bubble size
        label=species,
        alpha=0.7
    )
plt.title("Bubble Visualization: Sepal Dimensions and Area Proportion")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
plt.grid(True)
plt.show()



```

```python
# Task: Explore one advanced statistical function in SciPy and demonstrate its use with an example.

# Sample Task: Calculate kurtosis and skewness of a dataset.

# Code Example:
from scipy import stats
import seaborn as sns
data = np.random.normal(0, 1, 1000)

kurtosis = stats.kurtosis(data)
skewness = stats.skew(data)
print(f"Kurtosis: {kurtosis}")
print(f"Skewness: {skewness}")

plt.figure(figsize=(10,6))
sns.histplot(data, kde=True)
plt.title("Distribution of Data")
plt.show()
```
