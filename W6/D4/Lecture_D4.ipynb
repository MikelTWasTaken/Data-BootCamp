```python
import scipy.stats as stats

# 1. Gaussian Distribution: Probability of an item weighing more than 55 grams
mu = 50  # Mean
sigma = 5  # Standard deviation
weight_threshold = 55

# CDF gives the probability of being LESS than 55, so we subtract from 1
prob_gaussian = 1 - stats.norm.cdf(weight_threshold, loc=mu, scale=sigma)
print(f"Probability of weight > 55 grams: {prob_gaussian}")

# 2. Binomial Distribution: Probability of exactly 2 defective items in a batch of 10
n_trials = 10  # Number of items in a batch
p_success = 0.1  # Probability of a single defective item
k_defects = 2  # Exactly 2 defects

prob_binomial = stats.binom.pmf(k_defects, n=n_trials, p=p_success)
print(f"Probability of exactly 2 defective items: {prob_binomial}")

# 3. Poisson Distribution: Probability of 12 customers arriving in an hour
lambda_customers = 15  # Average rate of customer arrival
k_customers = 12  # Exactly 12 customers

prob_poisson = stats.poisson.pmf(k_customers, mu=lambda_customers)
print(f"Probability of exactly 12 customers arriving: {prob_poisson}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the Gaussian distribution
mu = 50  # Mean
sigma = 5  # Standard deviation

# Generate a range of values
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Compute the CDF
cdf_values = norm.cdf(x, loc=mu, scale=sigma)

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.plot(x, cdf_values, label='CDF', color='blue')
plt.title('CDF of Gaussian Distribution', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Cumulative Probability', fontsize=14)
plt.grid(True)
plt.legend(fontsize=12)
plt.show()
```

```python
from scipy.stats import binom

# Parameters for the Binomial distribution
n_trials = 10  # Number of trials
p_success = 0.5  # Probability of success in each trial

# Generate a range of possible outcomes
x = np.arange(0, n_trials + 1)

# Compute the CDF
cdf_values = binom.cdf(x, n=n_trials, p=p_success)

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(x, cdf_values, label='CDF', color='green', where='post')
plt.title('CDF of Binomial Distribution', fontsize=16)
plt.xlabel('Number of Successes', fontsize=14)
plt.ylabel('Cumulative Probability', fontsize=14)
plt.grid(True)
plt.legend(fontsize=12)
plt.show()
```


### **Scenario: Bakery Analysis**

A bakery wants to analyze customer purchases and ingredient variability. Here’s the breakdown:

1. **PMF**: The bakery sells different types of pastries: croissants, muffins, and scones. You want to calculate the probability of selling each type based on past data.
2. **PDF**: The weight of a croissant is normally distributed. You want to calculate the probability of a croissant weighing within a specific range.
3. **Gaussian Distribution**: Use the normal distribution to understand how consistent the croissant weights are.
4. **Binomial Distribution**: On average, 80% of customers buy croissants. Out of 10 customers, what’s the probability that exactly 8 of them buy croissants?

---

### **Given Data**:

1. **PMF Data**:
   - Croissants: Sold 40 times.
   - Muffins: Sold 35 times.
   - Scones: Sold 25 times.
   - Total: 100 sales.

2. **PDF Data**:
   - Average weight of a croissant: \( \mu = 60 \) grams.
   - Standard deviation: \( \sigma = 5 \) grams.

3. **Binomial Distribution**:
   - Probability of buying a croissant (\( p \)): 0.8.
   - Number of customers (\( n \)): 10.

---

### **Questions**:

1. What is the probability of selling each type of pastry (PMF)?
2. What is the probability that a randomly chosen croissant weighs between 58 and 62 grams (PDF)?
3. What is the probability that exactly 8 out of 10 customers buy croissants (Binomial)?



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, binom

# 1. PMF: Probability of selling each pastry type
pastry_sales = {"Croissants": 40, "Muffins": 35, "Scones": 25}
total_sales = sum(pastry_sales.values())
pmf = {key: value / total_sales for key, value in pastry_sales.items()}

print("PMF (Probability of Selling Each Pastry):")
for pastry, probability in pmf.items():
    print(f"{pastry}: {probability:.2f}")

# Visualization of PMF
plt.figure(figsize=(8, 5))
plt.bar(pmf.keys(), pmf.values(), color=['gold', 'blue', 'green'], alpha=0.7)
plt.title("PMF: Probability of Selling Each Pastry", fontsize=16)
plt.xlabel("Pastry Type", fontsize=14)
plt.ylabel("Probability", fontsize=14)
plt.ylim(0, 0.5)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

# 2. PDF: Probability of a croissant weighing between 58 and 62 grams
mu = 60  # Mean weight
sigma = 5  # Standard deviation
lower_bound = 58
upper_bound = 62

# Calculate cumulative probabilities
p_58 = norm.cdf(lower_bound, loc=mu, scale=sigma)
p_62 = norm.cdf(upper_bound, loc=mu, scale=sigma)
prob_weight_range = p_62 - p_58

print(f"\nProbability of a croissant weighing between 58 and 62 grams: {prob_weight_range:.2f}")

# Visualization of PDF
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000)
pdf = norm.pdf(x, loc=mu, scale=sigma)
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label='PDF', color='red')
plt.axvspan(lower_bound, upper_bound, color='blue', alpha=0.3, label='58-62 grams')
plt.title("PDF: Croissant Weights", fontsize=16)
plt.xlabel("Weight (grams)", fontsize=14)
plt.ylabel("Density", fontsize=14)
plt.legend(fontsize=12)
plt.grid()
plt.show()

# 3. Binomial: Probability of 8 out of 10 customers buying croissants
p_croissant = 0.8  # Probability of buying croissant
n_customers = 10  # Total customers
k = 8  # Exactly 8 customers

prob_binomial = binom.pmf(k, n=n_customers, p=p_croissant)
print(f"\nProbability of exactly 8 out of 10 customers buying croissants: {prob_binomial:.2f}")

# Visualization of Binomial Distribution
x_binom = range(0, n_customers + 1)
binomial_pmf = binom.pmf(x_binom, n=n_customers, p=p_croissant)
plt.figure(figsize=(8, 5))
plt.bar(x_binom, binomial_pmf, color='purple', alpha=0.7)
plt.title("Binomial PMF: Customer Croissant Purchases", fontsize=16)
plt.xlabel("Number of Customers Buying Croissants", fontsize=14)
plt.ylabel("Probability", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
