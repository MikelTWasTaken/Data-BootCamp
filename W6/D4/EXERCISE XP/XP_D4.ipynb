```python
# Objective: Apply linear regression to a dataset and interpret the results.

# Task: Given a dataset of housing prices (house_prices) and their corresponding sizes (house_sizes), use linear regression to predict the price of a house given its size.
import numpy as np
from scipy.stats import linregress
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
# What is the slope and intercept of the regression line?
# Predict the price of a house that is 90 square meters.
# Interpret the meaning of the slope in the context of housing prices.

house_sizes = np.array([50, 70, 80, 100, 120])
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

slope, intercept, r_value, p_value, std_err = linregress(house_sizes, house_prices)
print('The slope is', slope.round(2))
print('The intercept is', intercept.round(2))


# Predict the price of a house that is 90 square meters.
house_size = 90
predicted_price = (slope * house_size) + intercept
print('For a house of 90m, the price is approx: $',predicted_price.round(2))

plt.scatter(house_sizes, house_prices)
plt.plot(house_sizes, slope * house_sizes + intercept, color='springgreen')
plt.show()

# Analysis:
# For every 1 square meter increase in house size, the price increases by approximately 1828.77 dollars.Whereas 64383.56 (the intercept) is the predicted price of a house when its size is 0 square meters. While this value is not meaningful in a real-world housing context (as houses cannot have 0 square meters), it is necessary for the linear regression equation to model the relationship.
```

```python
# Objective: Test understanding of ANOVA and its application.

# Task: Three different fertilizers are applied to three separate groups of plants to test their effectiveness. The growth in centimeters is recorded.
# Dataset:
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]
# Questions:
# Perform an ANOVA test on the given data. What are the F-value and P-value?
# Based on the P-value, do the fertilizers have significantly different effects on plant growth?
    #with a small p-value of 2.7542672098048448e-05 At least one fertilizer has a statistically significant different effect on crop yield.
# Explain what would happen if the P-value were greater than 0.05.
    #There would be no indication that the fertilizers are different. 

f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print("F-value:", f_value)
print("P-value:", p_value)

data = [fertilizer_1, fertilizer_2, fertilizer_3]
plt.boxplot(data, labels=["Fertilizer 1", "Fertilizer 2", "Fertilizer 3"])
plt.title("ANOVA")
plt.xlabel("Fertilizer")
plt.ylabel("Growth (cm)")
plt.show()
```

```python
# Instructions
# Work with a binomial distribution: calculate probabilities for different numbers of successes.
# Example: Calculating the probability of getting exactly 5 heads in 10 coin flips.
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

n = 10
p = 0.5

binom_dist = stats.binom(n, p)
prob_5_successes = binom_dist.pmf(5)
print("PMF at 5 successes: %",(prob_5_successes*100).round(2))

x_binom = np.arange(0, n + 1)
pmf_binom = binom_dist.pmf(x_binom)

plt.figure(figsize=(6, 4))
plt.bar(x_binom, pmf_binom, alpha=0.7, color='blue', edgecolor='black')
plt.title('Binomial Distribution\n$n=10$, $p=0.5$')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Instructions
# Calculate the Pearson and Spearman correlation coefficients between two variables in a dataset.
# Example in code: data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]}) 
from scipy.stats import spearmanr, pearsonr
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})

age = data['age']
income = data['income']

correlation, p_value = spearmanr(age, income)
print("Spearman's correlation coefficient:", correlation)
print("p-value:", p_value)

correlation, p_value = pearsonr(age, income)
print("Pearson's correlation coefficient:", correlation)
print("p-value:", p_value)

# Analysis
# The spearman and pearson are close to 1 indicating age and income increase together. The p value is small in both cases meaning the relationship is significant.abs

plt.figure(figsize=(10, 6))
plt.scatter(age, income)
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Age vs. Income')
plt.show()
```
