```python
import matplotlib.pyplot as plt  

plt.plot([1, 2, 3], [4, 5, 6])  
plt.show()
```

```python
# Sample data
x = [1, 2, 3]
y = [4, 5, 6]

# Create the plot
plt.plot(x, y, color='blue', linestyle='--', marker='o', linewidth=2, label='Line 1')

# Customize title and labels
plt.title('Sample Plot', fontsize=14, color='darkblue', loc='center')
plt.xlabel('X-Axis Label', fontsize=12, color='black')
plt.ylabel('Y-Axis Label', fontsize=12, color='black')

# Customize grid
plt.grid(True, linestyle='--', color='gray', linewidth=0.5)

# Add a legend
plt.legend(loc='upper left', fontsize=10)

# Set axis limits
plt.xlim(0, 4)
plt.ylim(0, 7)

# Show the plot
plt.show()
```
