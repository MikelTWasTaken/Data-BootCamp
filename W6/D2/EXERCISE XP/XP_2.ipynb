```python
# ðŸŒŸ Exercise 6 : Basic Hypothesis Testing
# Instructions
# Create a sample dataset to test the effectiveness of a new training program on employee productivity:


import numpy as np
from scipy.stats import ttest_rel
import pandas as pd

# Productivity scores of employees before the training program
productivity_before = np.random.normal(loc=50, scale=10, size=30)

# Productivity scores of the same employees after the training program
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

# Your task is to formulate a hypothesis regarding the training program's effectiveness 
# and test it using basic statistical functions in NumPy. Given a dataset, formulate a simple hypothesis and test it using basic statistical functions in NumPy.

# HYPOTHESIS: The training program has a positive effect on productivity scores, with a mean improvement of at least 10 points.
# Null Hypothesis (H0): The mean productivity before the training program is equal to or less than the mean productivity after the training program.
# Alternative Hypothesis (H1): The mean productivity after the training program is greater than the mean productivity before the training program.  

# Calculate the mean productivity scores before and after the training program
mean_before = np.mean(productivity_before)

# Example of using NumPy to calculate the mean and standard deviation of productivity scores before and after the training program
mean_before = np.mean(productivity_before)
std_before = np.std(productivity_before)
mean_after = np.mean(productivity_after)
std_after = np.std(productivity_after)
print("Mean productivity before:", mean_before)
print("Standard deviation of productivity before:", std_before)
print("Mean productivity after:", mean_after)
print("Standard deviation of productivity after:", std_after)
# Example of using NumPy to calculate the t-test statistic and p-value
t_statistic, p_value = ttest_rel(productivity_before, productivity_after)
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

```python
# ðŸŒŸ Exercise 7 : Complex Array Comparison
# Instructions
# Create two arrays and perform element-wise comparison to find which elements are greater in the first array.

a = np.arange(12).reshape(3, 4)
print(a)

b = np.arange(12).reshape(4, 3).T
print(b)

# The expected output is a boolean array showing which elements in the first array are greater than the corresponding elements in the second array.
a_compare = a < b
print(a_compare)
print(type(a_compare))
print(a_compare.dtype)
```

```python
# ðŸŒŸ Exercise 8 : Time Series Data Manipulation
# Instructions
# Generate time series data for the year 2023. Demonstrate slicing for the following intervals:

# January to March
# April to June
# July to September
# October to December
# Generate a time series data for a specific period and demonstrate how to slice this data for different intervals.

jan_mar = np.arange('2023-01-01', '2023-04-01', dtype='datetime64[D]')
apr_jun = np.arange('2023-04-01', '2023-07-01', dtype='datetime64[D]')
jul_sep = np.arange('2023-07-01', '2023-10-01', dtype='datetime64[D]')
oct_dec = np.arange('2023-10-01', '2024-01-01', dtype='datetime64[D]')

first_week_jan = jan_mar[:7]  
mid_may = apr_jun[30:45]      
august_full = jul_sep[31:62]  
november_start = oct_dec[:30]

print("First week of January:", first_week_jan)
print("15 days in the middle of May:", mid_may)
print("All days in August:", august_full)
print("First 30 days of Q4:", november_start)
```
